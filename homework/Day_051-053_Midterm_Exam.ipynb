# <strong> Part1. 訓練資料

```python
import warnings
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

%matplotlib inline

plt.style.use("fivethirtyeight")
warnings.filterwarnings('ignore')
```

* ## <strong> 載入資料

```python
# Raw data
data = pd.read_csv("../midterm_data/train_data.csv")
data.head()
```

* ## <strong> 特徵理解

```python
# 檢查資料之數量(rows, columns)
print("Data shape: ", data.shape)
```

```python
# 檢查資料之型態、遺漏值、數量
data.info()
```

```python
# 檢查定量資料之統計值
data.describe()
```

```python
# 計算各欄位遺漏值數量
data.isnull().sum()
```

```python
# Drop 非特徵欄位及取得目標欄位
X = data.drop(['poi', 'email_address', 'name'], axis=1)
y = data['poi']
```

```python
print("Correlation with target:")
for column in X.columns:
    corr = data["poi"].corr(data[column])
    print("\t{}: {}".format(column, corr))
```

```python
# Drop 相關係數較低之欄位
X.drop(["to_messages"], axis=1, inplace=True)
```

```python
y.head()
```

```python
y = y.map({True:1, False:0})
```

# <strong> 二、 特徵改進: 清洗資料集

```python
# 特徵工程前先分割資料集
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.25, random_state=44)
```

```python
print("X train size: {}".format(X_train.shape))
print("X test size: {}".format(X_test.shape))
```

```python
# 計算遺漏值數量
print(X_train.isnull().sum())
print('-'*25)
print(X_test.isnull().sum())
```

```python
 X_train[['from_poi_to_this_person', 'from_messages', 'from_this_person_to_poi', 'shared_receipt_with_poi']].describe()
```

```python
# 遺漏值以 中位數 填補之欄位(偏態明顯)
for col in ['from_poi_to_this_person', 'from_messages', 'from_this_person_to_poi', 'shared_receipt_with_poi'] :
    X_train[col].fillna(X_train[col].median(), inplace=True)
    X_test[col].fillna(X_test[col].median(), inplace=True)
```

```python
df_isnull = X_train.isnull().sum()
non_zero_columns_train = list()
for column in df_isnull.index:
    if df_isnull[column] != 0:
        non_zero_columns_train.append(column)
```

```python
df_isnull = X_test.isnull().sum()
non_zero_columns_test = list()
for column in df_isnull.index:
    if df_isnull[column] != 0:
        non_zero_columns_test.append(column)
```

```python
# 其餘欄位遺漏值以 0 填補
for col in non_zero_columns_train:
    X_train[col].fillna(0, inplace=True)
for col in non_zero_columns_test:
    X_test[col].fillna(0, inplace=True)
```

```python
# 資料標準化: 平均值為0, 標準差為1
from sklearn.preprocessing import StandardScaler

scaler = StandardScaler()
X_train = scaler.fit_transform(X_train)
X_test = scaler.transform(X_test)
```

# <strong> 建立模型

```python
# 找出模型最佳參數組合
from sklearn.model_selection import GridSearchCV

def get_best_model_and_accuracy(model, params, X, y):
    grid = GridSearchCV(model, params, cv=5, verbose=1, scoring="accuracy")   
    grid.fit(X, y)
    print("Best accuracy: {}".format(grid.best_score_))
    print("Best parameters: {}".format(grid.best_params_))
    print("Average time(s) to fit: {}".format(round(grid.cv_results_["mean_fit_time"].mean(), 3)))
    print("Average time(s) to score: {}".format(round(grid.cv_results_["mean_score_time"].mean()), 3))
    means = grid.cv_results_['mean_test_score']
    stds = grid.cv_results_['std_test_score']
    #  這邊依設置的參數列印出所有參數的得分狀況
    for mean, std, params in zip(means, stds, grid.cv_results_['params']):
        print("%0.3f (+/-%0.03f) for %r" % (mean, std * 2, params))
    return grid
```

```python
from sklearn.linear_model import LogisticRegression
from sklearn.neighbors import KNeighborsClassifier
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestClassifier
from sklearn.ensemble import GradientBoostingClassifier
from sklearn.metrics import accuracy_score, f1_score, classification_report

lr_params = {"C": [0.1, 1, 10], "penalty": ["l1", "l2"]}
knn_params = {"n_neighbors": [1, 2, 3, 4, 5, 6, 7]}
tree_params = {"max_depth": [None, 1, 3, 5, 7], "min_samples_split": [2, 3, 4, 5], "min_samples_leaf": [1, 2, 3, 4]}
forest_params = {"n_estimators": [10, 50, 100, 150, 200], "max_depth": [None, 1 ,3, 5, 7], "min_samples_split": [2, 3, 4, 5], "min_samples_leaf": [1, 2, 3, 4]}
gradient_params = {"learning_rate": [0.1, 0.2, 0.3, 0.4], "n_estimators": [10, 50, 100, 150, 200], "max_depth": [None, 1, 3, 5, 7]}

lr = LogisticRegression()
knn = KNeighborsClassifier()
tree = DecisionTreeClassifier()
forest = RandomForestClassifier()
gradient = GradientBoostingClassifier()
```

```python
# LogisticRegression
lr = get_best_model_and_accuracy(lr, lr_params, X_train, y_train)

y_pred_lr = lr.predict(X_test)
print("Accuracy score: {}".format(accuracy_score(y_test, y_pred_lr)))
print("F1 score: {}".format(f1_score(y_test, y_pred_lr)))
print("Classification Report: {}".format(y_test, y_pred_lr))
```

```python
# KNeighborsClassifier
knn = get_best_model_and_accuracy(knn, knn_params, X_train, y_train)

y_pred_knn = knn.predict(X_test)
print("Accuracy score: {}".format(accuracy_score(y_test, y_pred_knn)))
print("F1 score: {}".format(f1_score(y_test, y_pred_knn)))
print("Classification Report: {}".format(y_test, y_pred_knn))
```

```python
# DecisionTreeClassifier
tree = get_best_model_and_accuracy(tree, tree_params, X_train, y_train)

y_pred_tree = tree.predict(X_test)
print("Accuracy score: {}".format(accuracy_score(y_test, y_pred_tree)))
print("F1 score: {}".format(f1_score(y_test, y_pred_tree)))
print("Classification Report: {}".format(y_test, y_pred_tree))
```

```python
# RandomRorestClassifier
forest = get_best_model_and_accuracy(forest, forest_params, X_train, y_train)

y_pred_forest = forest.predict(X_test)
print("Accuracy score: {}".format(accuracy_score(y_test, y_pred_forest)))
print("F1 score: {}".format(f1_score(y_test, y_pred_forest)))
print("Classification Report: {}".format(y_test, y_pred_forest))
```

```python
# GradientBoostingClassifier
gradient = get_best_model_and_accuracy(gradient, gradient_params, X_train, y_train)

y_pred_gradient = gradient.predict(X_test)
print("Accuracy score: {}".format(accuracy_score(y_test, y_pred_gradient)))
print("F1 score: {}".format(f1_score(y_test, y_pred_gradient)))
print("Classification Report: {}".format(y_test, y_pred_gradient))
```

```python
from mlxtend.classifier import StackingClassifier

meta_estimator = GradientBoostingClassifier(tol=100, subsample=0.7, n_estimators=100, max_depth=2, max_features="sqrt", learning_rate=0.2)
stacking = StackingClassifier(classifiers=[knn, lr, tree, forest, gradient], meta_classifier=meta_estimator, use_probas=True, average_probas=False)
stacking.fit(X_train, y_train)
y_pred_stacking = stacking.predict_proba(X_test)[:, 1]
y_pred_stacking
```

# <strong> 測試資料

* ## <strong> 特徵工程

```python
# 載入測試資料
data_test = pd.read_csv("../midterm_data/test_features.csv")

# 移除非特徵值
X_test = data_test.drop(['name', 'email_address', 'to_messages'], axis=1)

# 遺漏值以 中位數 填補之欄位(偏態明顯)
for col in ['from_poi_to_this_person', 'from_messages', 'from_this_person_to_poi', 'shared_receipt_with_poi'] :
    X_test[col].fillna(X_test[col].median(), inplace=True)
    
df_isnull = X_test.isnull().sum()
non_zero_columns = list()
for column in df_isnull.index:
    if df_isnull[column] != 0:
        non_zero_columns.append(column)

# 其餘欄位遺漏值以 0 填補
for col in non_zero_columns:
    X_test[col].fillna(0, inplace=True)

# 資料標準化: 平均值為0, 標準差為1
scaler = StandardScaler()
X_test = scaler.fit_transform(X_test)
```

* ## <strong> 模型預測

```python
y_pred_lr = lr.predict_proba(X_test)[:, 1]
y_pred_knn = lr.predict_proba(X_test)[:, 1]
y_pred_tree = tree.predict_proba(X_test)[:, 1]
y_pred_forest = forest.predict_proba(X_test)[:, 1]
y_pred_gradient = gradient.predict_proba(X_test)[:, 1]
y_pred_stacking = stacking.predict_proba(X_test)[:, 1]
```

* ## <strong> 儲存結果

```python
name = data_test['name']
```

```python
for model_name, model in zip(['lr', 'knn', 'tree', 'forest', 'gradient', 'stacking'], [y_pred_lr, y_pred_knn, y_pred_tree, y_pred_forest, y_pred_gradient, y_pred_stacking]):
    sub = pd.DataFrame({'name': name, 'poi': model})
    sub.to_csv('../midterm_data/mid-term_submission_{}.csv'.format(model_name), index=False)
```
